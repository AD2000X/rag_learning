vector.dot(vector) calculates the dot product of a vector with itself. A result close to 1 indicates that this is a normalized vector. A dot product close to 1 shows that the vector has been normalized to unit length, which is common in text embeddings, allowing texts of different lengths to be reasonably compared for similarity.

Why use the dot product?
The dot product is important in vector similarity comparisons because:

When vectors are normalized, the dot product is equivalent to cosine similarity.
It has a lower computational cost.
It effectively measures the directional similarity between two vectors.
In this example, the dot product of a vector with itself being close to 1 indicates that it is a normalized unit vector, which makes subsequent similarity comparisons more accurate and efficient.

What is a dot product?
The dot product is the sum of the products of corresponding elements in two vectors. For example, given two vectors A = [1,2,3] and B = [4,5,6] , the dot product is calculated as: (1×4) + (2×5) + (3×6) = 4 + 10 + 18 = 32

Dot Product Results:
Positive: Indicates that the vectors point in similar directions.
Negative: Indicates that the vectors point in opposite directions.
Zero: Indicates that the vectors are perpendicular.
The dot product is commonly used to calculate vector similarity, especially for comparing semantic similarity in text search.

***The question_text_vector is created by combining the question and answer (text) together, then converting it into a vector representation.

Example:
Original Data:
Question: "How to install Python?"
Answer: "First, download the Python installer, then run it and follow the instructions to complete the installation."

Processing Steps:
Combine Text: "How to install Python? First, download the Python installer, then run it and follow the instructions to complete the installation."
Convert to a 384-dimensional vector

Advantages:
Preserves the semantic relationship between the question and answer.
Understands the overall context of the question and answer.
Allows matching both similar questions and relevant answers during search.

Application:
When a user asks a question, this approach can retrieve results that match not only the question but also relevant answers.

Helps in understanding the semantic connection between question-answer pairs.
Improves search accuracy.

This is more comprehensive than using only the question vector or answer vector because it considers the semantic link between the question and answer.

```python
index_settings = {
    "settings": {
        "number_of_shards": 1,
        "number_of_replicas": 0
    },
    "mappings": {
        "properties": {
            "text": {"type": "text"},
            "section": {"type": "text"},
            "question": {"type": "text"},
            "course": {"type": "keyword"},
            "id": {"type": "keyword"},
            "question_vector": {          # embeddings for question
                "type": "dense_vector",
                "dims": 384,
                "index": True,
                "similarity": "cosine"
            },
            "text_vector": {              # embeddings for answer
                "type": "dense_vector",
                "dims": 384,
                "index": True,
                "similarity": "cosine"
            },
            "question_text_vector": {     # embeddings for question and answer***
                "type": "dense_vector",
                "dims": 384,
                "index": True,
                "similarity": "cosine"
            },
        }
    }
}

index_name = "course-questions"

es_client.indices.delete(index=index_name, ignore_unavailable=True)
es_client.indices.create(index=index_name, body=index_settings)
```

```python
from tqdm.auto import tqdm
```

# Data preparation

Generating Vector Representations:

Convert the question, answer, and combined text into vectors.
Add these vectors to each document.

```python
for doc in tqdm(documents):
    question = doc['question']
    text = doc['text']
    qt = question + ' ' + text

    doc['question_vector'] = model.encode(question)
    doc['text_vector'] = model.encode(text)
    doc['question_text_vector'] = model.encode(qt)
```

# Data storage

Store the Processed Documents (with Vectors) in Elasticsearch:

Write the data into the previously created index.
Create a searchable database.

```python
for doc in tqdm(documents):
    es_client.index(index=index_name, document=doc)
```

```python
def elastic_search_knn(field, vector, course):
    knn = {
        "field": field,
        "query_vector": vector,
        "k": 5,
        "num_candidates": 10000,
        "filter": {
            "term": {
                "course": course
            }
        }
    }

    search_query = {
        "knn": knn,
        "_source": ["text", "section", "question", "course", "id"]
    }

    es_results = es_client.search(
        index=index_name,
        body=search_query
    )

    result_docs = []

    for hit in es_results['hits']['hits']:
        result_docs.append(hit['_source'])

    return result_docs
```

```python
def question_vector_knn(q):
    question = q['question']
    course = q['course']
    doc_id=q['document']

    v_q = model.encode(question)

    return elastic_search_knn('question_vector', v_q, course)
```

```python
def hit_rate(relevance_total):
    cnt = 0

    for line in relevance_total:
        if True in line:
            cnt = cnt + 1

    return cnt / len(relevance_total)
```

```python
def mrr(relevance_total):
    total_score = 0.0

    for line in relevance_total:
        for rank in range(len(line)):
            if line[rank] == True:
                total_score = total_score + 1 / (rank + 1)

    return total_score / len(relevance_total)
```

```python
def evaluate(ground_truth, search_function):
    relevance_total = []

    for q in tqdm(ground_truth):
        doc_id = q['document']
        results = search_function(q)
        relevance = [d['id'] == doc_id for d in results]
        relevance_total.append(relevance)

    return {
        'hit_rate': hit_rate(relevance_total),
        'mrr': mrr(relevance_total),
    }
```

# elastic_search_knn_combined evaluate method:

Combines the similarity of three vectors (question_vector, text_vector, question_text_vector).

Uses cosineSimilarity to calculate the similarity for each vector and sums them up.

Filters by specific course as well.

This approach is likely more accurate because:

It considers the similarity of the question, answer, and combined text simultaneously.

Provides a comprehensive match by integrating multiple perspectives.

Results are more thorough.

```python
def elastic_search_knn_combined(vector, course):
    search_query = {
        "size": 5,
        "query": {
            "bool": {
                "must": [
                    {
                        "script_score": {
                            "query": {
                                "term": {
                                    "course": course
                                }
                            },
                            "script": {
                                "source": """
                                    cosineSimilarity(params.query_vector, 'question_vector') +
                                    cosineSimilarity(params.query_vector, 'text_vector') +
                                    cosineSimilarity(params.query_vector, 'question_text_vector') +
                                    1
                                """,
                                "params": {
                                    "query_vector": vector
                                }
                            }
                        }
                    }
                ],
                "filter": {
                    "term": {
                        "course": course
                    }
                }
            }
        },
        "_source": ["text", "section", "question", "course", "id"]
    }

    es_results = es_client.search(
        index=index_name,
        body=search_query
    )

    result_docs = []

    for hit in es_results['hits']['hits']:
        result_docs.append(hit['_source'])

    return result_docs
```

```python
def vector_combined_knn(q):
    question = q['question']
    course = q['course']

    v_q = model.encode(question)

    return elastic_search_knn_combined(v_q, course)

evaluate(ground_truth, vector_combined_knn)
```
